```python
lst1=[1,2,3,4]
lst2=[5,6,7,8]
result=map(lambda x, y: x+y, lst1, lst2)
print("addition of both the lists is:")
print(list(result))

num=[1,2,3,4,5]
def square(x):
  return x*x
result=list(map(square, num))
print(result)
```
