```python
print("half pyramid pattern of stars(*):")
n = int(input("enter the number of rows:"))
for i in range(n):
  for j in range(i+1):
    print("* ", end="")
  print()
```

```python
rows = int(input("enter the number of rows you want:"))
number=1
print("floyd's triangle:")
for i in range(1,rows+1):
  for j in range(1, i+1):
    print(number, end=" ")
    number=number+1
  print()
```
