```python
def check(words):
  c=0
  lst=[]
  for i in words:
    if len(i)>1 and i[0]==i[-1]:
      c+=1
      lst.append(i)
  print("list of words with first and last character same \n", lst)
  return c
count=check(["abc","cfc","xyz","aba","1221"])
print("number of words having first and last character same:", count)
```
