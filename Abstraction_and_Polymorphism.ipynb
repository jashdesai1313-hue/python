```python
from abc import ABC, abstractmethod
class Abcsclass(ABC):
  def print(self, x):
    print("passed value", x)
  @abstractmethod
  def task(self):
    print("we are inside abc class")
class test_class(Abcsclass):
  def task(self):
    print("we are inside test class")
test_obj = test_class()
test_obj.task()
test_obj.print(100)
```
