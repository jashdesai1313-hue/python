```python
def total_cal(bill_amount, tip_percentage):
  total=bill_amount*(1+tip_percentage/100)
  total=round(total, 2)
  print(f"your total bill is Rs. {total}")

total_cal(4365, 200)
```

```python
def divis_three(number):
  if number%3==0:
    return cube(number)
  else:
    return False

def cube(number):
  return number**3
print(divis_three(9))
print(divis_three(10))
print(cube (6))
```
