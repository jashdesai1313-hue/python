```python
st={1, 4 , 13, 24, 45}
print(st)
st={2.4, "jash", 46, 2, 4, (1, 2, 3, 4)}
print(st)
st={1, 2, 3, 4, 3, 2, 1, 4}
print(st)
st=set([1, 3, 2, 2])
print(st)
st.pop()
print(st)
```

```python
st1={"green", "blue"}
st2={"blue", "yellow"}
print(st1)
print(st2)
print("intersection of both the sets is:")
st3=st1.intersection(st2)
print(st3)
```
