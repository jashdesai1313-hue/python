```python
def pali(x):
  a=len(x)-1
  s=0
  while (s<a):
    if(x[s]!=x[a]):
      return False
    s+=1
    a-=1
  return True
t=(1, 2, 3, 3, 2, 1)
if pali(t):
  print("the tuple is a palindrome")
else:
  print("the tuple is not a palindrome")
```
